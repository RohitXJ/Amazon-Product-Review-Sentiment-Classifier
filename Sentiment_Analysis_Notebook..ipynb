# Clustering

## Text Processing and Vectorize

```python
def clean_text(text):
    text = text.lower()                                 # Lowercase
    text = re.sub(r"http\S+|www\S+|https\S+", '', text) # Remove URLs
    text = re.sub(r'\@w+|\#','', text)                  # Remove mentions and hashtags
    text = re.sub(r'[^A-Za-z\s]', '', text)             # Remove numbers & punctuation
    text = re.sub(r'\s+', ' ', text).strip()            # Remove extra spaces
    return text
    
df["clean_text"] = df["Text"].apply(clean_text)
df['clean_text']
```

```python
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
X = vectorizer.fit_transform(df["clean_text"])
```

## Applying KMeans Clustering

```python
inertia = []

K_range = range(2, 10)
for k in K_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X)
    inertia.append(km.inertia_)

plt.plot(K_range, inertia, marker='o')
plt.title("Elbow Curve for Optimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

### Conclusion: Optimal Number of Clusters
Best K = 4, based on the elbow method

Silhouette score not needed — it’ll likely support the same range, and you’re right to skip it due to time/resource costs

```python
km = KMeans(n_clusters=4, random_state=42)
df["cluster"] = km.fit_predict(X).astype('int32')
```

```python
df.info()
```

## Cluster Interpretation — What’s in each group?

```python
# 1. WordCloud for each cluster
for i in range(4):
    cluster_text = df[df['cluster'] == i]['clean_text'].str.cat(sep=' ')
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(cluster_text)
    plt.figure(figsize=(10, 4))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"Top Words in Cluster {i}")
    plt.show()
```

```python
# 2. Sample reviews from each cluster
for i in range(4):
    print(f"\n--- Cluster {i} Sample Reviews ---")
    print(df[df['cluster'] == i][['Score', 'Summary', 'clean_text']].sample(3, random_state=42).to_string())
```

```python
# 3. Avg score and review length by cluster
cluster_summary = df.groupby('cluster')[['Score', 'review_length', 'word_count']].mean().round(2)
print("\nAverage Stats by Cluster:")
print(cluster_summary)
```

# Classification

```python
def label_sentiment(score):
    if score in [1, 2]:
        return 0  # Negative
    elif score == 3:
        return 1  # Neutral
    else:
        return 2  # Positive

df["sentiment"] = df["Score"].apply(label_sentiment)
y = df["sentiment"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,stratify=y, random_state=42)
```

```python
def Eval(model,name):
    y_pred = model.predict(X_test)
    
    print(f"Classification Report for {name}:")
    print(classification_report(y_test, y_pred, target_names=["Negative", "Neutral", "Positive"]))
    
    print(f"Confusion Matrix for {name}:")
    print(confusion_matrix(y_test, y_pred))
```

## Logistic Regression

```python
clf = LogisticRegression(max_iter=10000, class_weight='balanced', random_state=42)
clf.fit(X_train, y_train)
Eval(clf, 'Logistic Regression')
```

## Random Forest Classifier

```python
rf_model = RandomForestClassifier(class_weight='balanced', random_state=42, n_jobs=-1,n_estimators=10)
rf_model.fit(X_train, y_train)

Eval(rf_model, "Random Forest")
```

## Linear Support Vector Classifier (LinearSVC)

```python
svc_model = LinearSVC(class_weight='balanced', random_state=42)
svc_model.fit(X_train, y_train)

Eval(svc_model, "Linear SVC")
```

## XGBoost Classifier (with weight handling via scale_pos_weight)

```python
xgb_model = XGBClassifier(objective='multi:softmax', num_class=3, use_label_encoder=False,
                          eval_metric='mlogloss', random_state=42, n_jobs=-1)

xgb_model.fit(X_train, y_train)

Eval(xgb_model, "XGBoost Classifier")
```

## 🔧 Model Selection Summary

We selected **LinearSVC** and **XGBoost Classifier** for the following reasons:

### ✅ LinearSVC

* Achieved **84% accuracy** and strong macro F1.
* Offers **fast training time** and **efficient memory usage**.
* Performs consistently across all classes, especially **Positive and Negative**.
* Great balance of **speed vs. performance**; ideal for tuning and production.

### ⚖️ XGBoost (GPU)

* Achieved **85% accuracy** with **exceptional Positive class recall**.
* GPU-accelerated: suitable for large-scale experimentation.
* Although Neutral class recall is low, it shows **potential with hyperparameter tuning**.
* Selected for its **flexibility and scalability** in optimized environments.

These models combine efficiency and power, allowing us to optimize for both speed and accuracy.


# Hyperparameter Tuning

## LinearSVC

```python
param_grid_svc = {
    'C': [0.1, 1, 10],
    'loss': ['hinge', 'squared_hinge'],
    'class_weight': ['balanced']
}

svc = LinearSVC(max_iter=7500)
grid_svc = GridSearchCV(svc, param_grid_svc, cv=3, verbose=1, n_jobs=-1)
grid_svc.fit(X_train, y_train)

best_svc = grid_svc.best_estimator_
Eval(best_svc, "Tuned LinearSVC")
```

## XGBoost with GPU

```python
param_grid_xgb = {
    'max_depth': [6, 10],
    'learning_rate': [0.1, 0.2],
    'n_estimators': [100, 200],
    'subsample': [0.8, 1.0],
    'scale_pos_weight': [1, 2],
    'tree_method': ['gpu_hist']  # GPU training
}

xgb = XGBClassifier(objective='multi:softmax', num_class=3, use_label_encoder=False, eval_metric='mlogloss')
grid_xgb = GridSearchCV(xgb, param_grid_xgb, cv=3, verbose=1)
grid_xgb.fit(X_train, y_train)

best_xgb = grid_xgb.best_estimator_
Eval(best_xgb, "Tuned XGBoost (GPU)")
```
